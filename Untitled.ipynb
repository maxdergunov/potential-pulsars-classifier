```python
import pandas as pd
import numpy as np

from sklearn.datasets import make_classification
from sklearn.decomposition import PCA

from matplotlib import pyplot as plt 
import seaborn as sns

from imblearn.over_sampling import SMOTE, ADASYN, RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler, NearMiss, CondensedNearestNeighbour
```

```python
X, y = make_classification(n_samples=10000, n_features=2, n_informative=2, n_redundant=0, n_repeated=0, n_classes=2, n_clusters_per_class=1, weights=[0.1,0.9], flip_y=0.01, class_sep=1.2, hypercube=True, shift=0.0, scale=1.0, shuffle=True, random_state=29)
```

```python
X
```

```python
df = pd.DataFrame(X)
#df['class'] = y

ax = sns.scatterplot(data=df)
```

```python
plt.figure(figsize=(18, 18))

plt.subplot(321)
plt.title('Data')
X1, Y1 = X, y
plt.scatter(X1[:, 0], X1[:, 1], marker='o', c=Y1,
            s=25, edgecolor='k')

```

```python
undersampler = RandomUnderSampler(random_state=25)

X_rus, y_rus = undersampler.fit_resample(X, y)

plt.figure(figsize=(18, 18))

plt.subplot(321)
plt.title('RUS')
X1, Y1 = X_rus, y_rus
plt.scatter(X1[:, 0], X1[:, 1], marker='o', c=Y1,
            s=25, edgecolor='k')
```

```python
undersampler = NearMiss(version=1, n_jobs=-1)

X_nm1, y_nm1 = undersampler.fit_resample(X, y)

plt.figure(figsize=(18, 18))

plt.subplot(321)
plt.title('NM1')
X1, Y1 = X_nm1, y_nm1
plt.scatter(X1[:, 0], X1[:, 1], marker='o', c=Y1,
            s=25, edgecolor='k')
```

```python
undersampler = NearMiss(version=2, n_jobs=-1)

X_nm2, y_nm2 = undersampler.fit_resample(X, y)

plt.figure(figsize=(18, 18))

plt.subplot(321)
plt.title('NM2')
X1, Y1 = X_nm2, y_nm2
plt.scatter(X1[:, 0], X1[:, 1], marker='o', c=Y1,
            s=25, edgecolor='k')
```

```python
undersampler = NearMiss(version=3, n_jobs=-1)

X_nm3, y_nm3 = undersampler.fit_resample(X, y)

fig = plt.figure(figsize=(18, 18))
ax = fig.add_subplot(1,1,1)


plt.subplot(321)
plt.title('NM3')
X1, Y1 = X_nm3, y_nm3
plt.scatter(X1[:, 0], X1[:, 1], marker='o', c=Y1,
            s=25, edgecolor='k')

ax.legend()
plt.show()
```

```python
f, (ax1, ax2, ax3) = plt.subplots(3, 1, sharey=True, figsize=(6, 14))

ax1.scatter(X_nm1[:, 0], X_nm1[:, 1], marker='o', c=y_nm1,
            s=25, edgecolor='k')
ax1.set_title('NM1')

ax2.scatter(X_nm2[:, 0], X_nm2[:, 1], marker='o', c=y_nm2,
            s=25, edgecolor='k')
ax2.set_title('NM2')

ax3.scatter(X_nm3[:, 0], X_nm3[:, 1], marker='o', c=y_nm3,
            s=25, edgecolor='k')
ax3.set_title('NM3')

```

```python
df = pd.read_csv('./pulsar_stars.csv')

train_df = df.sample(frac=0.7,random_state=25)
test_df = df.drop(train_df.index)

X_train = train_df.drop('target_class', axis=1)
y_train = train_df['target_class']

#print('Starting NearMiss-3 at', datetime.now().strftime("%H:%M:%S"))
undersampler = NearMiss(sampling_strategy='auto', version=3, n_jobs=-1)
X_train_nm3, y_train_nm3 = undersampler.fit_resample(X_train, y_train)
#print('Done at', datetime.now().strftime("%H:%M:%S"))

pca = PCA(n_components=2, random_state=25)

X_train_nm3 = pca.fit_transform(X_train_nm3)

plt.figure(figsize=(18, 18))

plt.subplot(321)
plt.title('NM3')
X1, Y1 = X_train_nm3, y_train_nm3
plt.scatter(X1[:, 0], X1[:, 1], marker='o', c=Y1,
            s=25, edgecolor='k')

ax.legend()
plt.show()

```

```python
undersampler = CondensedNearestNeighbour(sampling_strategy='auto', n_jobs=-1)

X_cnn, y_cnn = undersampler.fit_resample(X, y)

fig = plt.figure(figsize=(18, 18))
ax = fig.add_subplot(1,1,1)


plt.subplot(321)
plt.title('CNN')
X1, Y1 = X_cnn, y_cnn
plt.scatter(X1[:, 0], X1[:, 1], marker='o', c=Y1,
            s=25, edgecolor='k')


plt.show()
```

```python
oversampler = RandomOverSampler(sampling_strategy='auto')

X_ros, y_ros = oversampler.fit_resample(X, y)

fig = plt.figure(figsize=(18, 18))
ax = fig.add_subplot(1,1,1)


plt.subplot(321)
plt.title('ROS')
X1, Y1 = X_ros, y_ros
plt.scatter(X1[:, 0], X1[:, 1], marker='o', c=Y1,
            s=25, edgecolor='k')


plt.show()
```

```python
oversampler = SMOTE(sampling_strategy='auto', random_state=25)

X_smote, y_smote = oversampler.fit_resample(X, y)

fig = plt.figure(figsize=(18, 18))
ax = fig.add_subplot(1,1,1)


plt.subplot(321)
plt.title('SMOTE')
X1, Y1 = X_smote, y_smote
plt.scatter(X1[:, 0], X1[:, 1], marker='o', c=Y1,
            s=25, edgecolor='k')


plt.show()
```

```python
oversampler = ADASYN(sampling_strategy='auto', random_state=25)

X_adasyn, y_adasyn = oversampler.fit_resample(X, y)

fig = plt.figure(figsize=(18, 18))
ax = fig.add_subplot(1,1,1)


plt.subplot(321)
plt.title('ADASYN')
X1, Y1 = X_adasyn, y_adasyn
plt.scatter(X1[:, 0], X1[:, 1], marker='o', c=Y1,
            s=25, edgecolor='k')


plt.show()

fig.savefig('./images/adasyn.png')
```
